#### 0. import essential modules

```python
import pandas as pd
```

#### 1. Import CSV data file, and make column 'date' datetime format (5% + 5%)

```python
# define filename
fn_data = 'Rain_MD-Androy-Bekily_1d_19810101-20210530_CHIRPS.csv'

# read csv file, a dataframe will be created to store data
df_rain_1d = pd.read_csv(fn_data)
```

```python
# Convert 'str' to 'datetime', so we can do some time series analysis operations
df_rain_1d['date'] = pd.to_datetime(df_rain_1d['date'], format='%Y/%m/%d')
```

```python
df_rain_1d.head(5)
```

#### 2. Calculate rainfall totals for each month, and export to a CSV file named _Rain_MD-Androy-Bekily_1M_19810101-20210530_CHIRPS.csv_ (25% + 5%)

```python
df_rain_1m = df_rain_1d.resample('1M', on='date').sum()
df_rain_1m.head(5)
```

```python
fn_data_out_1m_sum = 'Rain_MD-Androy-Bekily_1M_19810101-20210530_CHIRPS.csv'
df_rain_1m.to_csv(fn_data_out_1m_sum, float_format='%.2f', date_format='%Y/%m')
```

#### 3. Calculate maximum 1-day rainfall for each month, and export to a CSV file named _Rain_MD-Androy-Bekily_1d-1M-Max_19810101-20210530_CHIRPS.csv_ (25% + 5%)

```python
df_rain_1m_max = df_rain_1d.resample('1M', on='date').max()
df_rain_1m_max.head(5)
```

```python
fn_data_out_1m_max = 'Rain_MD-Androy-Bekily_1d-1M-Max_19810101-20210530_CHIRPS.csv'
df_rain_1m_max.to_csv(fn_data_out_1m_max, float_format='%.2f', date_format='%Y/%m', index=False)
```

#### 4. Calculate maximum 3 calendar-day rainfall for calendar month December for each year, and export to a CSV file named _Rain_MD-Androy-Bekily_3d-Dec-Max_19810101-20210530_CHIRPS.csv_ (25% + 5%)

```python
df_rain_3d = df_rain_1d.resample('3D', on='date').sum()
df_rain_3d.head(5)
```

```python
df_rain_3d_max_1m = df_rain_3d.resample('1M').max()
df_rain_3d_max_1m.head(5)
```

```python
df_rain_3d_max_Dec = df_rain_3d_max_1m[df_rain_3d_max_1m.index.month == 12]
df_rain_3d_max_Dec.head(5)
```

```python
fn_data_out_3d_max_dec = 'Rain_MD-Androy-Bekily_3d-Dec-Max_19810101-20210530_CHIRPS.csv'
df_rain_3d_max_Dec.to_csv(fn_data_out_3d_max_dec, float_format='%.2f', date_format='%Y/%m')
```

It has provided quasi-global rainfall estimate for over 35 years. You can find more information about CHRIPS via this link: https://www.nature.com/articles/sdata201566
